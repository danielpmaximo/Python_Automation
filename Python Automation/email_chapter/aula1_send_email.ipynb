```python
import smtplib

conn = smtplib.SMTP('smtp.gmail.com', 587)
type(conn)
```

```python
conn.ehlo() # Check if the connection is secure
```

```python
conn.starttls() # Upgrade the connection to a secure encrypted SSL/TLS connection
```

```python
conn.login('daniel.pessoa.maximo@gmail.com', 'isyu ceey xkni byrb')
```

```python
conn.sendmail('daniel.pessoa.maximo@gmail.com', # Sender's email
              'daniel.pessoa.maximo@gmail.com', # Recipient's email
              'Subject: Test\n\nThis is a test email sent from Python!') # Send the email
```
