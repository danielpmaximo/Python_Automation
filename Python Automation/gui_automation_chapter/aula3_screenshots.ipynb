```python
import pyautogui
```

```python
pyautogui.screenshot('screenshot.png')
```

```python
pyautogui.locateOnScreen('test.png', confidence=0.8) # This will return the coordinates of the image if found, or None if not found
```

```python
pyautogui.locateCenterOnScreen('test.png', confidence=0.8) # This will return the center coordinates of the image if found, or None if not found
```

```python
pyautogui.moveTo(777, 142, duration=1) # Moves the mouse to the specified coordinates (777, 142) over a duration of 1 second
```
